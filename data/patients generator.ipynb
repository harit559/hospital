```python
from mimesis import Personal, Datetime
from mimesis.enums import Gender
```

```python
generator_person = Personal()
generator_dt = Datetime()
```

```python
patients = []
for i in range(1000000):
    if i%2:
        sex = Gender.FEMALE
    else:
        sex = Gender.MALE
    date = generator_dt.date(1900, 2017)
    month, day, year = date.split('/')
    pat = {
        'name': generator_person.name(sex),
        'surname': generator_person.surname(sex),
        'day': day,
        'month': month,
        'year': year,
        'weight': generator_person.weight(),
        'height': generator_person.height(),
        'blood_type': generator_person.blood_type(),

    }
    patients.append(pat)
    
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(patients)
```

```python
df
```

```python
df = df[['name','surname','day','month','year','height','weight','blood_type']]
```

```python
df.to_csv('patients.csv',index=False, header=False)
```
